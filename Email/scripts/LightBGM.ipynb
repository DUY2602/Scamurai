```python
# Building the LightBGM Model
import numpy as np
import pickle
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Load the matrix from models into the LightBGM
X = np.load("../models/X_features.npy")
y_raw = np.load("../models/y_labels.npy", allow_pickle=True)

# Use the number translated album that XGBoost created
with open("../models/label_encoder.pkl", "rb") as f:
    le = pickle.load(f)
y = le.transform(y_raw)

# Separate the data into 70-30 percentage, 70 for model training and 30 for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Training LightBGM process
lgb_model = lgb.LGBMClassifier(n_estimators=300, learning_rate=0.1, random_state=42, n_jobs=-1)
lgb_model.fit(X_train, y_train)

# Testing the LightBGM 
y_pred = lgb_model.predict(X_test)
print(f"\n Result: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Save the LightBGM Model into models
with open("../models/lgb_model.pkl", "wb") as f: 
    pickle.dump(lgb_model, f)
print("Saved succesfully into models")
```
