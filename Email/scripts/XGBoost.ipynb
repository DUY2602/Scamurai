```python
# Building the XGBoost Model
import numpy as np
import pickle
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Load the matrix from models into the XGBoost
X = np.load("../models/X_features.npy")
y_raw = np.load("../models/y_labels.npy", allow_pickle=True)

# Translate the label into numbers and save it
le = LabelEncoder()
y = le.fit_transform(y_raw) 
with open("../models/label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

# Separate the data into 2 parts, 70% for training and 30% for testing (Train and Test Data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Training the XGBoost
xgb_model = xgb.XGBClassifier(n_estimators=300, learning_rate=0.1, random_state=42, n_jobs=-1)
xgb_model.fit(X_train, y_train)

# Test the XGBoost 
y_pred = xgb_model.predict(X_test)
print(f"\nResult: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Save the result into models
with open("../models/xgb_model.pkl", "wb") as f: 
    pickle.dump(xgb_model, f)
print("Saved successfully into models")
```
